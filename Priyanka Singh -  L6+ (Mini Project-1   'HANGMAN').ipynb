```python
import random
file=['python','India','China','Japan','Russia','Covid','Corona','Vaccine']
```

```python
def body(tries):
    step=[ """
            -----------
            |           |
            |           O
            |         --|--    
            |          //\\
            |            
            |           
           ---                  
                             """,
          
          """
            -----------
            |           |
            |           O
            |         --|--   
            |          /   
            |            
            |           
           ---               
                             """    ,
          
          """
            -----------
            |           |
            |           O
            |         --|-- 
            |            
            |            
            |           
           ---           
                                  """,
          
          """
            -----------
            |           |
            |           O
            |         --| 
            |            
            |            
            |            
           ---          
                                """,
          
          
          """
            ------------
            |           |
            |           O
            |           | 
            |            
            |            
            |
           ---           
                                  """,
          
          """
            --------------
            |           |
            |           O
            |            
            |            
            |            
            |           
           ---        
                                  """,
          
          
         """
            ------------
            |           |
            |           
            |            
            |            
            |            
            |           
           ---        
                                  """,
          """
            ------------
            |           
            |           
            |            
            |            
            |            
            |           
           ---        
                                  """
         ]
    return step[tries]
```

```python
def guess_word():
    word=random.choice(file).upper()
    blank='-'*len(word)
    
    print('Lets Play HANGMAN GAME!')
    print('There are ',len(word),'letters in the word')
    guessed=False
    guessed_letter=[]
    tries=7
    print(body(tries))
    print(blank)
    while not guessed and tries>=0:
        guess=input('Guess a letter: ').upper()
        if len(guess)==1 and guess.isalpha():
            if guess in guessed_letter:
                print('You have already guessed a letter')
            elif guess not in word:
                print('\n')
                print('Oops that a incorrect guess')
                
                tries-=1 
                if tries<7 and tries>2:
                    print('You are left with',tries,'tries')
                elif tries==2:
                    print('You are left with',tries,'tries')
                    print('This is getting really close')
                elif tries==1:  
                    print('You are left with',tries,'try')
                    print('This is getting really really  close')
                elif tries==0:
                    print('You are left with',tries,'try')
                    print('One more wrong guess and you could lose...!!')
                
                print(body(tries))  
                print(blank)
                guessed_letter.append(guess)
            else:
                print('Yuppie you have guessed it correct')
                print('You have',tries,'Tries left with you')
                print(body(tries))
                wordlist=list(blank)
                guessed_letter.append(guess)
                value=[i for i, letter in enumerate(word) if letter==guess]
                for i in value:
                    wordlist[i]=guess
                blank=''.join(wordlist)
                print(blank)
                print('\n')
                if '-' not in blank:
                    guessed=True
                    
            
        else:
            print('Not a valid guess, please guess a alphabetic letter')
            print('\n')
            print(body(tries))
            
    
    if guessed:
        print('Conragts You won!!,the word is ',word)
        
    else:
        print('You loose!! the word was',word, 'better luck next time')
                
guess_word()
```
