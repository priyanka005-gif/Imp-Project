```python
l1=[' ']*10
l1[0]='#'
def show():
    print(l1[1]," | ",l1[2]," | ",l1[3])
    print('--------------')
    print(l1[4]," | ",l1[5]," | ",l1[6])
    print('--------------')
    print(l1[7]," | ",l1[8]," | ",l1[9])
    
```

```python
def select(option):
    return ((l1[1]==l1[2]==l1[3]==option) or (l1[4]==l1[5]==l1[6]==option) or (l1[7]==l1[8]==l1[9]==option) or
           (l1[1]==l1[4]==l1[7]==option) or (l1[2]==l1[5]==l1[8]==option) or (l1[3]==l1[6]==l1[9]==option) or 
           (l1[1]==l1[5]==l1[9]==option) or (l1[3]==l1[5]==l1[7]==option))
     
```

```python
def open(position):
    return l1[position]==' '
```
